```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df = pd.read_csv('/home/user/BirthOfSSL/mieb_evals/Model_Benchmark_Results.csv')
df = df.set_index('model name')
df.head()
```

```python
avg_df = df.mean(axis=1)
var_df = df.var(axis=1)

avg_df.plot(kind='bar')
plt.show()

var_df.plot(kind='bar')
plt.show()

plt.figure(figsize=(8,6))
plt.scatter(avg_df, var_df)
plt.xlabel('Average Accuracy')
plt.ylabel('Variance')
plt.title('Variance vs Average Accuracy per Model')
plt.grid(True)
plt.show()
```

```python

avg_df.to_csv('/home/user/BirthOfSSL/mieb_evals/Model_Benchmark_Avg.csv', header=True)
var_df.to_csv('/home/user/BirthOfSSL/mieb_evals/Model_Benchmark_Var.csv', header=True)
```
